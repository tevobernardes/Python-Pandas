```python
import pandas as pd

# Leitura dos arquivos.

df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
# Juntando todos os arquivos.

df = pd.concat([df1, df2, df3, df4, df5])

df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
# Alterando o tipo de dado da coluna LojaID.

df["LojaID"]= df["LojaID"].astype("object")
```

```python
# Consultando valores nulos das colunas.

df.isnull().sum()
```

```python
# Criando uma nova tabela de Receita Valor de venda X Quantidade. 

df["Receita"] = df["Vendas"].mul(df["Qtde"])

df.head()
```

```python
# Ex: Criando uma nova tabela de Receita|Vendas para saber a quantidade quando não a tem.

df["Receita|Vendas"] = df["Receita"] / df["Vendas"]

df.head()
```

```python
# Maior Receita

df["Receita"].max()
```

```python
# Menor Receita

df["Receita"].min()
```

```python
# Maior Receita e/ou Vendas com top de 5

df.nlargest(5, "Receita")
```

```python
# Menor Receita e/ou Vendas com top de 3

df.nsmallest(3, "Receita")
```

```python
# Agrupamento de Cidades com Total da Receita.

df.groupby("Cidade")["Receita"].sum()
```

```python
# Ordenando o conjunto de dado Receita do Maior para o Menor.

df.sort_values("Receita", ascending=False)

```
