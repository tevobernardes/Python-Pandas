```python
# Importando a biblioteca pandas e efetuando a leitura do arquivo.

import pandas as pd

df = pd.read_excel("/Fortaleza.xlsx")
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Alterando o tipo de dado de data para inteiro.

df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
# Alterando o tipo de dado quando temos uma tabela de data Ex:

df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Total de vendas pelo ano.

df.groupby(df["Data"].dt.year)["Vendas"].sum()
```

```python
# Criando uma coluna somente com o ano de venda.

df["Ano|Vendas"] = df["Data"].dt.year

# Pegando uma amostra.

df.sample()
```

```python
# Extraindo mês e o dia

df["Mês|Vendas"], df["Dia|Vendas"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample()
```

```python
# Retornando a data mais antiga

df["Data"].min()
```

```python
# Calculando a diferença de dias

df["Diferença|Dias"] = df["Data"] - df["Data"].min()
df.sample()
```

```python
# Trimestre

df["Trimestre|Vendas"] = df["Data"].dt.quarter
df.sample()
```

```python
# Filtrando as vendas de 2019 do mês de fevereiro

vendas_Fevereiro = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 2)]
```
