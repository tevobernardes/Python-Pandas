```python
import pandas as pd
```

```python
# Abrindo Arquivo CSV com elimição de erros (error_bad_lines=False) e tirando o ; (sep=";")

df = pd.read_csv("/Gapminder.csv",error_bad_lines=False, sep=";")
```

```python
# Exibindo as 5 primeiras linhas por padrão.

df.head()
```

```python
# Renomeando os nomes das colunas.

df = df.rename(columns={"country":"País", "continent":"Continente", "year":"Ano", "lifeExp":"Expectativa de Vida", "pop":"População", "gdpPercap":"PIB"})
```

```python
df.head(10)
```

```python
# Total de linhas e colunas.

df.shape
```

```python
# Exibindo os nomes das colunas.

df.columns
```

```python
# Exibindo os tipos de dados de cada coluna.

df.dtypes
```

```python
# Exibindo as 5 últimas linhas por padrão.

df.tail()
```

```python
# Exibindo informações estatísticas.

df.describe()
```

```python
# Filtrando | Exibindo valores unicos da tabela continente

df["Continente"].unique()
```

```python
# Localizando (loc) dentro da tabela Continente somente a Oceania.

oceania = df.loc[df["Continente"] == "Oceania"]
oceania.head()
```

```python
oceania["Continente"].unique()
```

```python
# Agrupando todos os Continentes com a quantidades de países.

df.groupby("Continente")["País"].nunique()
```

```python
# Exibindo a média da expectativa de vida e ano total.

media = df.groupby("Ano")["Expectativa de Vida"].mean()
print(media)
```

```python
# Soma total do PIB.

df["PIB"].sum()
```
